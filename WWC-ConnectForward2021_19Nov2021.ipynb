## General Steps in a Data Analysis Project:
    1. Collect/Aquire the dataset
    2. Get the dataset in required file format - csv, json, xlsx etc.
    3. Load the dataset into a pandas dataframe 
    4. Data Wrangling - 
       Filter/Drop any extra/non-essential rows/columns
       Handle Missing values in remaining dataset
       Check for datatypes, formats of the data in your dataframe
    5. Exploratory Analysis of your data - statistical, visual
    6. Feature Engineering, Normalization/Standardization/Scaling
    7. Training a Model for predictions   
    8. Drawing insights

## Grouping Data - 
Pandas has groupby(), pivot_table() and crosstab() functions for grouping and summarizing data in a dataframe. 

# Pivot table

Pivot tables help to arrange and rearrange (or "pivot") statistics in order to draw attention to useful information. 
This speeds up the data analysis and the search for useful information and better decision making. (Wikipedia)

The pivot table takes simple column-wise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data. 

```python
# Pivot table 
example_df = pd.DataFrame({"Name": ["Anita", "Rohan", "Ram", "Geeta"],
                           "Age": [24, 34, 23,43],
                          "Gender": ['F', 'M','M','F'],
                          "Salary": [20000, 34000, 29000, 30000]})
example_df.head()
```

```python
example_df.pivot_table(index = ['Gender', 'Name'], values = 'Salary', aggfunc = 'sum')
```

```python
example_df.pivot_table(index = ['Gender'], values = 'Age', aggfunc = 'mean')
```

```python
example_df.pivot_table(index = ['Gender'], values = 'Age', columns = 'Name', aggfunc = 'mean', fill_value = 0)
```

```python
example_df.pivot_table(index = ['Gender'], values = 'Salary',  aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
```

```python
pt = example_df.pivot_table(index = ['Gender'], values = 'Salary',  aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
pt.query('Gender == ["F"]')
```

## Group_by() does similar aggregation over a few dimensions

# cross tab
The crosstab() in pandas computes a frequency table of values in columns. We can specify columns and 
the aggregation function to be applied on the columns.


```python
## Printing a crosstab for df
pd.crosstab(example_df.Name, example_df.Age)
```

```python
pd.crosstab(example_df.Gender, example_df.Age, values = 'Age', aggfunc='sum')          
```
